```python
import tensorflow as tf
from tensorflow.keras import layers, Sequential
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
```

```python
#paraametry modelu i treningu
max_features = 10000
max_len = 500
batch_size = 32
embedding_dims = 50
units=32
epochs=5
```

```python
#załadowanie do danych do zbiorów
(x_train,y_train),(x_test,y_test) = imdb.load_data(num_words=max_features)
```

```python
#dostaowanie recenzji do stałej dlugości
x_train = sequence.pad_sequences(x_train,maxlen=max_len)
x_test = sequence.pad_sequences(x_test,maxlen=max_len)
```

```python
model = Sequential([
    layers.Embedding(max_features,embedding_dims,input_length=max_len),
    layers.LSTM(units),
    layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,
          batch_size=batch_size,
          epochs=epochs,
          validation_data=(x_test,y_test))
```

```python
loss,accuracy=model.evaluate(x_test,y_test,batch_size=batch_size)
print(f"test accuracy: {accuracy}")
```

```python
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
recenzje = [
    "This movie was fantastic! I loved every moment of it.",
    "Terrible acting and boring plot.Waste of time.",
    "The storyline was okay but the acting was superb."
]
```

```python
#przetwarzanie recenzji do formatu zgodnego z modelem
word_index = imdb.get_word_index()
sequences = [[word_index[word] for word in review.split() if word in word_index] for review in recenzje]
sequences = pad_sequences(sequences,maxlen=max_len)
```

```python
predictions = model.predict(sequences)
```

```python
#klasyfikacja sentymentu
sentiment_labels = ['Negative','Positive']
for i, prediction in enumerate(predictions):
    sentiment = sentiment_labels[int(np.round(prediction))]
    print(f'recenzja: {recenzje[i]}')
    print(f'ocena sentymentu: {sentiment}\n')
```
